```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
import os as os
%matplotlib inline
```

```python
# https://jupyter.readthedocs.io/en/latest/projects/jupyter-directories.html
#print(os.getcwd())
```

```python
#df = pd.read_csv("../../output/data_analysis.csv")
df = pd.read_csv("../../output/data_analysis_cleaned.csv")
df[" E_Line"] = df[" E_Line"].astype(str).astype(int).div(1000)
df[" E_Module"] = df[" E_Module"].astype(str).astype(int).div(1000)
df[" NC"] = df[" NC"].div(1000)
df[" NO"] = df[" NO"].div(1000)
```

```python
NC = df[' NC']
NO = df[" NO"]
E_Line = df[" E_Line"]
E_Module = df[" E_Module"]
```

```python
NC_Max = NC.max()
NO_Max = NO.max()
E_Line_Max = E_Line.max() 
E_Module_Max = E_Module.max() 

#print(df.dtypes)

print("Max for NC: " + str(NC_Max))
print("Max for NO: " + str(NO_Max))
print("Max for E_Line: " + str(E_Line_Max))
print("Max for E_Module: " + str(E_Module_Max))
```

```python
sns.pairplot(df, x_vars=[' NC', ' NO'], y_vars=' E_Line', size=10, aspect=0.7, kind='reg')
sns.pairplot(df, x_vars=[' NC', ' NO'], y_vars=' E_Module', size=10, aspect=0.7, kind='reg')
```
