```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import os as os
%matplotlib inline
```

```python
print(os.getcwd())
```

```python
df = pd.read_csv("../../output/data_analysis.csv")
```

```python
NC = df[' NC']
NO = df[" NO"]
E_Line = df[" E_Line"]
```

```python
NC_Max = NC.max()
NO_Max = NO.max()
E_Line_Max = E_Line.max()

print("Max for NC: " + str(NC_Max))
print("Max for NO: " + str(NO_Max))
print("Max for E_Line: " + str(E_Line_Max))
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot(NC, E_Line, '.')
ax.set_xlabel('NC')
ax.set_ylabel('E_Line')
ax.set_xlim(0., NC_Max)
ax.set_ylim(0., E_Line_Max)
```
