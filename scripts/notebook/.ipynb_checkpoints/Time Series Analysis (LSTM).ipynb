```python
# https://github.com/gianfelton/12-Month-Forecast-With-LSTM/blob/master/12-Month%20Forecast%20With%20LSTM.ipynb
# https://www.youtube.com/watch?v=h8v9--50mdU
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.tools.eval_measures import rmse
from sklearn.preprocessing import MinMaxScaler
from keras.preprocessing.sequence import TimeseriesGenerator
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../../output/data_analysis_major_versions_with_dates.csv", usecols=[' E_Module', ' Release Date'])
df.head()
```

```python
# https://towardsdatascience.com/time-series-analysis-visualization-forecasting-with-lstm-77a905180eba
df[' Release Date'] = pd.to_datetime(df[' Release Date'])
# df['Year'] = df[' Release Date'].apply(lambda x: x.year)
# df['Quarter'] = df[' Release Date'].apply(lambda x: x.quarter)
# df['Month'] = df[' Release Date'].apply(lambda x: x.month)

df = df.set_index(" Release Date")

df.head()
```

```python
# pd.pivot_table(df,index=["Year", "Quarter"],values=[" NC", " NO", " E_Module", " E_Line", " T_Module", " T_Line"], aggfunc=np.sum)
# pd.pivot_table(df,index=["Month"],values=[" NC", " NO", " E_Module", " E_Line", " T_Module", " T_Line"], aggfunc=np.sum)
```

```python
train, test = df[:-12], df[-12:]
```

```python
scaler = MinMaxScaler()
scaler.fit(train)
train = scaler.transform(train)
test = scaler.transform(test)
```

```python
n_input = 12
n_features = 1
generator = TimeseriesGenerator(train, train, length=n_input, batch_size=6)
```

```python
model = Sequential()
model.add(LSTM(200, activation='relu', input_shape=(n_input, n_features)))
model.add(Dropout(0.15))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit_generator(generator,epochs=90)
```

```python
pred_list = []

batch = train[-n_input:].reshape((1, n_input, n_features))

for i in range(n_input):   
    pred_list.append(model.predict(batch)[0]) 
    batch = np.append(batch[:,1:,:],[[pred_list[i]]],axis=1)
```

```python
df_predict = pd.DataFrame(scaler.inverse_transform(pred_list),
                          index=df[-n_input:].index, columns=['Prediction'])

df_test = pd.concat([df,df_predict], axis=1)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(df_test.index, df_test[' E_Module'])
plt.plot(df_test.index, df_test['Prediction'], color='r')
plt.legend(loc='best', fontsize='xx-large')
plt.xticks(fontsize=18)
plt.yticks(fontsize=16)
plt.show()
```

```python
pred_actual_rmse = rmse(df_test.iloc[-n_input:, [0]], df_test.iloc[-n_input:, [1]])
print("rmse: ", pred_actual_rmse)
```
