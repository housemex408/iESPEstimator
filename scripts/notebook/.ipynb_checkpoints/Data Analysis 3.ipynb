```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
import os as os
import statsmodels as sm
import statsmodels.api as smapi
import statsmodels.regression.linear_model as lm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
%matplotlib inline
```

```python
#df = pd.read_csv("../../output/data_analysis_part2.csv")
# This has the releases with zero NC/NO removed
# Need to investigate to see why this happens in all -rc1 releases
df = pd.read_csv("../../output/data_analysis_major_versions.csv")
df[" E_Line"] = df[" E_Line"].astype(str).astype(int)
df[" E_Module"] = df[" E_Module"].astype(str).astype(int)
df[" NC"] = df[" NC"]
df[" NO"] = df[" NO"]
```

```python
df.describe()

def calculate_PRED(percentage, dataFrame):
    countLessPercent = dataFrame[dataFrame['Percent Error'] < percentage]['Percent Error']
    # print(countLessPercent.count())
    # print(dataFrame['Percent Error'].count())
    pred = countLessPercent.count() / dataFrame['Percent Error'].count()
    return pred
#df.dtypes
```

```python
# print out a correlation matrix of our data frame
corr = df.corr()

# display it
display(corr)
```

```python
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdBu')
```

```python
df.plot.scatter(x=' NC', y=' E_Line', c='red')
df.plot.scatter(x=' NO', y=' E_Line', c='green')
df.plot.scatter(x=' T_Module', y=' E_Line', c='orange')
df.plot.scatter(x=' T_Line', y=' E_Line', c='orange')
```

```python
# Let's create multiple regression
X = df[[' NC', ' NO', ' T_Module']]
Y = df[' E_Line']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.75, random_state=0)

linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
#linear_regression.fit(X, Y)
LinearRegression(fit_intercept=True, copy_X=True, n_jobs=None, normalize=False)

y_pred = linear_regression.predict(X_test)

coeff_df = pd.DataFrame(linear_regression.coef_, X.columns, columns=['Coefficient'])
coeff_df

# Interpreting Restuls:  https://www.youtube.com/watch?v=U7D1h5bbpcs  
```

```python
data = {'Observed':y_test, 'Predicted':y_pred.round(2), 'Difference':abs(y_test - y_pred).round(2), 'Percent Error':(abs(y_test - y_pred)/y_test).round(2)}
results = pd.DataFrame(data) 
results['Version'] = np.arange(results.__len__())
results
```

```python
pred25 = calculate_PRED(0.25, results)
print("PRED (25): " + "{:.2%}".format(pred25))

pred50 = calculate_PRED(0.50, results)
print("PRED (50): " + "{:.2%}".format(pred50))
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
ax = results.plot(kind='scatter', x='Version', y='Observed', c='green', label='Observed')
results.plot(kind='scatter', x='Version', y='Predicted', c='blue', label='Predicted', ax=ax)
```
