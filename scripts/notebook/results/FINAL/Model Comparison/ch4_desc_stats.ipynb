```python
import numpy as np
import pandas as pd
import os
import sys
import seaborn as sb
import matplotlib.pyplot as plt 
sys.path.append(os.path.abspath("../../../../"))
import Constants as c
import Utilities as utils
%matplotlib inline
```

```python
results = "../../05_30_2020/model_performance_results.csv"
columns = [c.MODEL, c.R_SQUARED, c.MAE, c.PRED_25, c.PRED_50, c.T_RECORDS, c.T_RECORDS, "P_NA"]
df = pd.read_csv(results, usecols=columns)

filter1 = df[c.MODEL] != c.MODULE_UC
filter2 = df[c.MODEL] != c.MODULE_UC
filter3 = df[c.MODEL] != c.LINE_UC
filter4 = df[c.MODEL] != c.LINE_UC

df = df[filter1 & filter2 & filter3 & filter4]
pd.options.display.float_format = "{:.2f}".format
df = utils.remove_outlier(df, c.MAE)
fig = plt.figure(figsize=(10,5))
plt.tight_layout()
sb.set(style="dark", palette="pastel")
ax = sb.boxplot(x=c.MODEL, y=c.MAE, data=df)
```
